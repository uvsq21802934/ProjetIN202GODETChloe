```python
def rang(lettre):
    return ord(lettre)-97
```

```python
def decalage(lettre_message,lettre_cle):
    if 97 <= ord(lettre_message) <= 122:
        return chr((rang(lettre_message) + rang(lettre_cle)) % 26 + 97)
    else:
        return lettre_message
```

J'ai ensuite simplement repris les fonctions rang et decalage vu plus haut, et fait en sorte que le mot de passe s'applique à tout le texte.
C'est-à-dire que j'ai fait une boucle, qui passe par chaque caractère (c), et qui enregistre la position du caractère dans le texte (i). Ainsi, la première lettre "d" est en position 0 dans le texte. A chaque position, j'affecte à une variable j, i%4 car mon mot de passe fait 4 lettres. Ainsi 0%4 = 0, la lettre du mot de passe par laquelle le caractère est codé sera en position 0 également. Et ainsi de suite, "c" le deuxième caractère donnera j = 1%4 = 1 donc la lettre du mot de passe par laquelle il est codé sera en position 1 donc l. 

J'obtiens ainsi le texte totalement décodé.

```python
def rang(lettre):
    return ord(lettre)-97


def decalage2(lettre_message,lettre_cle):
    if 97 <= ord(lettre_message) <= 122:
        return chr((rang(lettre_message) - rang(lettre_cle)) % 26 + 97)
    else:
        return lettre_message

texte_3 = "dceuq e n'ehfp cg p'kyhhep uqfw cgiy citudm c gzudiq ni ezhd px c jhptv ep cggsht. kg hdtymdt xdzei gdx rzyq wir mvzxpw, cifcchdb znwd ccyw wy lkcsht, dp isgd uqfw wy ?"
mot_de_passe = "clez"

texte_3_decrypte = ""
for i, c in enumerate(texte_3):
    j = i%4
    lettre_cle = mot_de_passe[j]
    texte_3_decrypte += decalage2(c, lettre_cle)
     
print(texte_3_decrypte)
```
